```python
cat_order =['never', 'less1', '1~3', '4~8', 'gt8']
ordinal_cat = ["Restaurant20To50" , "RestaurantLessThan20", "CarryAway", "CoffeeHouse", "Bar", "age", "income"]
income_map =[
 'Less than $12500',
 '$12500 - $24999',
 '$25000 - $37499',
 '$37500 - $49999',
 '$50000 - $62499',
 '$62500 - $74999',
 '$75000 - $87499',
 '$87500 - $99999',
 '$100000 or More'
]

age_order =['below21', '21', '26', '31', '36', '41', '46', '50plus']

```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
ord_cat_processor = Pipeline(steps = [
    ("missing_values",SimpleImputer(strategy = "most_frequent")),
    ("ordinal_encoding", OrdinalEncoder(categories= [cat_order, cat_order, cat_order, cat_order, cat_order, age_order, income_map]))
])
```

```python
cat_processor = Pipeline(steps = [
    ("cat", SimpleImputer(strategy = "most_frequent")),
    ("one_hot", OneHotEncoder(handle_unknown = "ignore", sparse_output = False))
])
```

```python
# df.drop(["Y", "education", "occupation"], axis = 1)
```

```python
cat = ['destination', 'passanger', 'weather', 'time', 'coupon', 'expiration',
       'gender', 'maritalStatus']
```

```python
preprocessing = ColumnTransformer(transformers = [
    ("ord", ord_cat_processor,ordinal_cat),
    ("cat", cat_processor, cat),
    ("num_rescale", StandardScaler(), ["temperature"])
], remainder = "passthrough")
```

```python
model = Pipeline(steps = [
    ("Preprocessing", preprocessing),
    ("Model", RandomForestClassifier())
])
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(["Y", "education", "occupation", "car"], axis = 1), df["Y"], test_size =.2, random_state = 42)
```

```python
X_train.head()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
